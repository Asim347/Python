### USD to PKR Converter Program

```python
US_value = float(input("Enter value in US dollars: "))
PKR_value = float(input("Enter the current exchange rate (PKR per 1 USD): "))
calculation = US_value * PKR_value
print(US_value, "USD at the rate of", PKR_value, "PKR/USD is equal to", calculation, "PKR")
```

### BMI Calculatio

```python
weight = float(input("Enter your weight in kilograms (kg): "))
height = float(input("Enter your height in meters (m): "))

bmi = weight / (height ** 2)
print(f"Your BMI is: {bmi:.2f}")
```

### Convert Temperature

```python
Fahrenheit = float(input("Enter temperature in Fahrenheit: "))
Celsius = (Fahrenheit - 32) * 5/9
print(f"Temperature in Celsius is: {Celsius:.2f}°C")
```

### Functions

```python
def check_num(num):
  if num % 2 == 0:
    print("The number is Even")
  else:
    print("The number is odd")


num = int(input("Enter any number to check whether it is odd or even"))
check_num(num)
```

### Find Maximum

```python
def maximum(num1, num2):
  if num1 > num2:
     print("num1 is greatest ")
  else:
    print("num2 is greatest")

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

maximum(num1, num2)
```

### Square of a number

```python
def square(num):
  squar = num * num
  print("Square of ", num ,"is", squar)

num = int(input("Enter any number to find the square :"))
square(num)
```
