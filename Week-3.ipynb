### Making Decisions

```python
def sum(a,b):
  print(a+b)

def sub(a,b):
  print(a+b)

def divide(a,b):
  print(a / b)

def multiple(a,b):
  print(a * b)

def remainder(a,b):
  print(a % b)

def calculator():
  a = int(input("Enter first value : "))
  b = int(input("Enter second value : "))
  op = input("Enter +,-,*,%,/ operator ")

  if op == '+':
    print(sum(a,b))
  elif op == '-':
    print(sub(a,b))
  elif op == '/':
    print(divide(a , b))

  elif op == '*':
    print(multiple(a,b))
  elif op == '%':
    print(remainder(a , b))
  else:
    print("Invalid operator")

calculator()
```

```python
def open_door(is_student, is_teacher, course):
    relevant_courses = ['AI', 'CS']  # Fixed the course list

    if is_student == "True":
        if course in relevant_courses:
            print("Allow")
        else:
            print("Not allowed")
    elif is_teacher == "True":
        if course in relevant_courses:
            print("Allow")
        else:
            print("Not allowed")
    else:
        print("Not allowed")

is_student = input("Is this a student ")
is_teacher = input("Is this a teacher ")
course = input("Enter course for both student and teacher :")

open_door(is_student, is_teacher, course)
```
